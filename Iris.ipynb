```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
```

```python
# Convert dataset to a DataFrame
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# Display the first few rows
print(df.head())
```

Visualize the Relationship
Let’s see how Sepal Length affects Petal Length.
Use seaborn to print out a scatterplot based on the dataset

```python
# Scatter plot
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', data=df)
plt.title('Sepal Length vs. Petal Length')
plt.show()
```

You should see a positive trend – as sepal length increases, petal length increases too.

Apply Linear Regression
We’ll predict Petal Length from Sepal Length.

```python
# Prepare the data
X = df[['sepal length (cm)']]  # Input (must be 2D)
y = df['petal length (cm)']    # Output

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Get the slope (m) and intercept (b)
print(f"Slope (m): {model.coef_[0]}")
print(f"Intercept (b): {model.intercept_}")
```

So, the equation for our line is:

Petal Length = 1.85 × Sepal Length − 7.10


Make Predictions
Let’s predict the petal length for a flower with a sepal length of 6 cm.

𝑌 = 1.85 × 6 − 7.10


```python
prediction = model.predict([[6]])
print(f"Predicted Petal Length: {prediction[0]:.2f}")
```

```python
from sklearn.metrics import r2_score

# Predictions for all points
y_pred = model.predict(X)

# Evaluate the model
r2 = r2_score(y, y_pred)

print(f"R²: {r2:.3f}")
```

R² = 0.76 means 76% of the variation in petal length is explained by sepal length. Not bad!

Visualize the Line
Let’s plot our regression line on the scatterplot.

```python
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', data=df)
plt.plot(X, y_pred, color='red', linewidth=2)  # Regression line
plt.title('Linear Regression: Sepal Length vs. Petal Length')
plt.show()
```
